## Vote Choice

```python
import plotly.express as px

vote_bar = df_clean.vote.value_counts().reset_index()
vote_bar = df_clean.groupby('vote').agg({'vote': 'size', 'ftbiden1': 'mean', 'fttrump1': 'mean'})
vote_bar = vote_bar.rename(columns={'vote': 'num_votes'})
vote_bar['ftbiden1'] = round(vote_bar['ftbiden1'], 2)
vote_bar['fttrump1'] = round(vote_bar['fttrump1'], 2)
vote_bar = vote_bar.reset_index()
vote_bar['percent'] = round(100*vote_bar['num_votes']/sum(vote_bar['num_votes']),2)

vote_bar['text'] = vote_bar['percent'].astype(str) + '%'

fig = px.bar(vote_bar.sort_values(by='num_votes', ascending=False), x='vote', y='percent', color='vote',
             labels={'vote': 'Vote choice', 'percent': 'Percent'},
             title='Vote choice as of April 2020',
             hover_data=['num_votes', 'ftbiden1', 'fttrump1'],
             text='text')
fig.update_layout(showlegend=False)
fig.update(layout=dict(title=dict(x=0.5)))
fig.show()
```

According to the data, Donald Trump and Joe Biden were neck to neck as of April 2020.

```python
colpercent = round(100*pd.crosstab(df_clean.vote, df_clean.partyID, normalize='columns'),2).reset_index()
colpercent = pd.melt(colpercent, id_vars='vote', 
                     value_vars=['Democrat', 'Republican', 'independent']).rename(columns={'value': 'colpercent'})

rowpercent = round(100*pd.crosstab(df_clean.vote, df_clean.partyID, normalize='index'),2).reset_index()
rowpercent = pd.melt(rowpercent, id_vars='vote',
                     value_vars=['Democrat', 'Republican', 'independent']).rename(columns={'value': 'rowpercent'})

votes = pd.crosstab(df_clean.vote, df_clean.partyID).reset_index()
votes = pd.melt(votes, id_vars='vote', 
                value_vars=['Democrat', 'Republican', 'independent']).rename(columns={'value': 'num_votes'})

ftb = pd.crosstab(df_clean['vote'], df_clean['partyID'], values=df_clean['ftbiden1'], aggfunc='mean').round(2).reset_index()
ftb = pd.melt(ftb, id_vars='vote',
              value_vars=['Democrat', 'Republican', 'independent']).rename(columns={'value': 'Biden thermometer'})

ftt = pd.crosstab(df_clean['vote'], df_clean['partyID'], values=df_clean['fttrump1'], aggfunc='mean').round(2).reset_index()
ftt = pd.melt(ftt, id_vars='vote',
              value_vars=['Democrat', 'Republican', 'independent']).rename(columns={'value': 'Trump thermometer'})

df_groupbar = pd.merge(colpercent, rowpercent, on=['vote', 'partyID'], validate='one_to_one')
df_groupbar = pd.merge(df_groupbar, votes, on=['vote', 'partyID'], validate='one_to_one')
df_groupbar = pd.merge(df_groupbar, ftb, on=['vote', 'partyID'], validate='one_to_one')
df_groupbar = pd.merge(df_groupbar, ftt, on=['vote', 'partyID'], validate='one_to_one')

df_groupbar['coltext'] = df_groupbar['colpercent'].astype(str) + '%'
df_groupbar['rowtext'] = df_groupbar['rowpercent'].astype(str) + '%'
```

```python
fig = px.bar(df_groupbar, x='vote', y='colpercent', color='partyID', 
             labels={'vote': 'Vote choice', 'colpercent': 'Percent'},
             title = 'Vote choice as of April 2020',
             hover_data = ['num_votes', 'Biden thermometer', 'Trump thermometer'],
             text = 'coltext',
             barmode='group')
fig.update_layout(showlegend=True)
fig.update(layout=dict(title=dict(x=0.5)))
fig.show()
```

```python
fig = px.bar(df_groupbar, x='rowpercent', y='vote', color='partyID',
             labels={'vote': 'Vote choice', 'rowpercent': 'Percent'},
             title='Vote choice as of April 2020',
             hover_data=['num_votes', 'Biden thermometer', 'Trump thermometer'],
             text='rowtext',
             barmode='stack')
fig.update_layout(showlegend=True)
fig.update(layout=dict(title=dict(x=0.5)))
fig.show()
```

```python
fig = px.bar(df_groupbar, x='vote', y='colpercent', color='partyID', facet_col='partyID',
             labels={'vote': 'Vote choice', 'colpercent': 'Percent'},
             title='Vote choice as of April 2020',
             hover_data=['num_votes', 'Biden thermometer', 'Trump thermometer'],
             text='coltext')
fig.update_layout(showlegend=False)
fig.update(layout=dict(title=dict(x=0.5)))
fig.for_each_annotation(lambda a: a.update(text=a.text.replace('partyID=', '')))
fig.show()
```

```python
fig = px.bar(df_groupbar, x='partyID', y='rowpercent', color='partyID', 
             facet_col='vote', facet_col_wrap=2,
             labels={'partyID': 'Party identification', 'rowpercent': 'Percent'},
             title='Vote choice as of April 2020',
             hover_data=['num_votes', 'Biden thermometer', 'Trump thermometer'],
             text='rowtext')
fig.update_layout(showlegend=False)
fig.update(layout=dict(title=dict(x=0.5)))
fig.for_each_annotation(lambda a: a.update(text=a.text.replace('vote=', '')))
fig.show()
```

There is no surprise here, when you look at the numbers by party.

### Hypothesis tests

#### Whether there is significant difference between men and women on how they rate Trump vs. Biden.

```python
from scipy import stats

fttrump_men = df_clean.query("sex=='Male'").fttrump1.dropna()
fttrump_women = df_clean.query("sex=='Female'").fttrump1.dropna()

stats.ttest_ind(fttrump_men, fttrump_women, equal_var=False)
```

Therefore, we can reject the null hypothesis and conclude that there is a statistically significant difference between men and women in terms of how highly they rate Trump, on average.

```python
ftbiden_men = df_clean.query("sex=='Male'").ftbiden1.dropna()
ftbiden_women = df_clean.query("sex=='Female'").ftbiden1.dropna()

stats.ttest_ind(ftbiden_men, ftbiden_women, equal_var=False)
```

We failed to reject the null hypothesis and that women and men have the same average thermometer rating to Biden.

### Multidimensional Scaling

```python
from sklearn.metrics import euclidean_distances

df_ft = df_clean.loc[:, 'fttrump1': 'ftdemocraticparty'].dropna()
names = df_ft.columns.str[2:]
ft_distance = euclidean_distances(np.transpose(df_ft))
```

```python
from sklearn import manifold

mds = manifold.MDS(n_components=2, dissimilarity='precomputed')
mds_fit = mds.fit(ft_distance)
pos = mds_fit.embedding_

plt.rcParams['figure.figsize'] = 12,8
plt.scatter(pos[:, 0], pos[:, 1], color='navy', s=0)

for i in enumerate(names):
    plt.annotate(i[1], (pos[i[0], 0], pos[i[0], 1]))
plt.show();
```

### Multiple Correspondence Analysis (MCA)

Similar to PCA, but MCA works with categorical observed features instead of continuous-valued ones. Like multidimensional scaling, MCA can be used to generate a map that is an excellent tool for EDA. In this case, the map will show would closely categories of different features correspond to one another.

```python
from prince import MCA

df_cat = df_clean[['vote16', 'vote', 'confecon', 'self_ideology', 'immignum', 'health_care', 'worry_covid', 
                   'worry_covid_economy', 'partyID', 'sex', 'science', 'universal_income', 'freecol']].dropna()
mca = MCA(n_components=2)
mca = mca.fit(df_cat)

mca.plot_coordinates(X=df_cat, 
                     figsize=(12, 12), 
                     show_row_points=False, 
                     row_points_size=10, 
                     show_row_labels=False, 
                     show_column_points=True,
                     column_points_size=30,
                     show_column_labels=True,
                     legend_n_cols=1);
```

```python
mca.column_coordinates(df_cat).sort_values(0)
```

The first latent feature we have categories is for liberal ideology, voting for Hillary Clinton in 2016, identifying as a Democrat, voting for Joe Biden in 2020, favoring free college a great deal, and believing science is extremely important. On the other end we have categories for conservative ideology, opposing a great deal on both universal basic income and free college, identifying as a republican, no at all worrying about the economy. This first latent feature appears to be the left-right ideological dimension.

```python
mca.column_coordinates(df_cat).sort_values(1)
```

For the second latent feature, we have positions that do not take a position: Did not vote in 2016, probably will not vote in 2020, and neither favor nor oppose universal basic income, and free college. This appears to measure voter apathy.
